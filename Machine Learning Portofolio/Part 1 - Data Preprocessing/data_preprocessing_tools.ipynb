##Importing the dataset

```python
# Step 1: Load the dataset
dataset = pd.read_csv('Data.csv')

# Step 2: Create matrix of features (X) and target vector (y)
x = dataset.iloc[:, :-1].values  # All rows, all columns except last
y = dataset.iloc[:, -1].values   # All rows, only last column
```

```python
print(x)
```

```python
print(y)
```

##Taking care of missing data

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(x[:, 1:3])
x[:, 1:3] = imputer.transform(x[:, 1:3])
```

Step-by-Step Code Breakdown

"from sklearn.impute import SimpleImputer"

üìå Imports the SimpleImputer class from scikit-learn.

"imputer = SimpleImputer(missing_values=np.nan, strategy='mean')"

üß± Creates an instance of SimpleImputer.

missing_values=np.nan: Tells it what to look for (NaN).

strategy='mean': Tells it how to replace (with average).

"imputer.fit(X[:, 1:3])"

üìä Fits the imputer to the numeric columns only (columns 1 and 2, i.e., age and salary).

It calculates the mean for each column where NaN exists.

"X[:, 1:3] = imputer.transform(X[:, 1:3])"

üîÅ Transforms the data by replacing all NaN values with the calculated mean.

The updated values are stored back into the original matrix X, replacing old columns.

##Encoding categorical data

###Encoding the independent variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
x = np.array(ct.fit_transform(x))
```

```python
print(x)
```

###Encoding the dependent variable

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
print(y)
```

#Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

#Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train[:, 3:] = sc.fit_transform(x_train[:, 3:])
x_test[:, 3:] = sc.transform(x_test[:, 3:])
```
